```python
area1 = [(179, 575),(134, 537),(209, 483),(229, 537)] 
cap = cv2.VideoCapture("/Users/mayurpatowary/Desktop/Project/DCUTEST.mp4")
count = 0
p = 0
var1 = 0
var2 = 0
while True:
    ret, frame = cap.read()
    if not ret:
        break
    count += 1
    if count % 4 != 0:
        continue
    #frame = cv2.resize(frame, (1020,500))
    #frame = frame[0:500, 250: 600]
    results = model.predict(frame) 
    px = pd.DataFrame(results[0].boxes.data).astype("float") 
    list = []
    for index,row in px.iterrows():
        x1 = int(row[0]) 
        y1 = int(row[1]) 
        x2 = int(row[2]) 
        y2 = int(row[3]) 
        d = int(row[5])
        c = class_list[d] 
        if 'person' in c:
            check1 = cv2.pointPolygonTest(np.array(area1,np.int32), ((x1, y2)), False) 
            if check1 >= 1:
                cv2.rectangle(frame,(x1,y1),(x2,y2), (0,255,0), 2)               
                cv2.circle(frame, (x1, y2), 4, (255,0,255), -1)               
                #cv2.putText(frame, "zone 1" , (x1,y1), cv2.FONT_HERSHEY_COMPLEX, (0.4), (255,255,255), 1) 
                playsound('/Users/mayurpatowary/Desktop/Project/Text to Speech/text_to_speech.mp3')
  
    cv2.polylines(frame,[np.array(area1,np.int32)], True, (255,0,0), 2)
    cv2.imshow("RGB", frame)
    if cv2.waitKey(1) & 0xFF == 27:
        break
cap.release()
cv2.destroyAllWindows()
cv2.waitKey(1)
```
